```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate ciccle
x,y=make_circles(n_samples=300,noise=0.1,factor=0.5,random_state=0)
# plot points with their labels
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.RdYlGn,edgecolors='k',marker='o',s=50)
plt.xlabel('Feature 1 (X[:,0])')
plt.ylabel('Feature 2 (X[:,1])')
plt.title('Circles Dataset')
plt.colorbar(label='Class Label')
plt.show()
```

# Circles Dataset Visualization

This notebook generates and visualizes a synthetic "circles" dataset using `scikit-learn` and `matplotlib`.

## Overview

The code uses the `make_circles` function from `sklearn.datasets` to create a dataset consisting of two interlocked circles. This type of dataset is often used to illustrate the challenges of linear separability and the need for non-linear models in machine learning.

## Dependencies

- `numpy`
- `matplotlib`
